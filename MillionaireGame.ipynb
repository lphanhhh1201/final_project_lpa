```python
import tkinter as tk
from tkinter import messagebox
from PIL import Image, ImageTk

class Quiz:
    def __init__(self):
        self.questions = [
            {"question": "Câu 1: Thủ đô của Việt Nam là?", "options": ["Hà Nội", "TP. Hồ Chí Minh", "Đà Nẵng", "Huế"], "answer": "Hà Nội"},
            {"question": "Câu 2: Con sông dài nhất thế giới?", "options": ["Sông Nile", "Sông Amazon", "Sông Mississippi", "Sông Hằng"], "answer": "Sông Nile"},
            {"question": "Câu 3: Kim tự tháp Giza nằm ở quốc gia nào?", "options": ["Ai Cập", "Mexico", "Peru", "Ấn Độ"], "answer": "Ai Cập"},
            {"question": "Câu 4: Thành phố lớn nhất nước Mỹ?", "options": ["New York", "Los Angeles", "Chicago", "Houston"], "answer": "New York"},
            {"question": "Câu 5: Tác giả của tiểu thuyết 'Những người khốn khổ'?", "options": ["Victor Hugo", "Leo Tolstoy", "Charles Dickens", "Fyodor Dostoevsky"], "answer": "Victor Hugo"},
            {"question": "Câu 6: Đơn vị tiền tệ của Nhật Bản?", "options": ["Yen", "Won", "Peso", "Ringgit"], "answer": "Yen"},
            {"question": "Câu 7: Đỉnh núi cao nhất thế giới?", "options": ["Everest", "K2", "Kangchenjunga", "Lhotse"], "answer": "Everest"},
            {"question": "Câu 8: Nước nào lớn nhất theo diện tích?", "options": ["Nga", "Canada", "Trung Quốc", "Hoa Kỳ"], "answer": "Nga"},
            {"question": "Câu 9: Quốc gia có dân số đông nhất thế giới?", "options": ["Trung Quốc", "Ấn Độ", "Hoa Kỳ", "Indonesia"], "answer": "Ấn Độ"},
            {"question": "Câu 10: Tác giả của bản giao hưởng số 9?", "options": ["Beethoven", "Mozart", "Bach", "Chopin"], "answer": "Beethoven"},
            {"question": "Câu 11: Thủ đô của Pháp?", "options": ["Paris", "Lyon", "Marseille", "Nice"], "answer": "Paris"},
            {"question": "Câu 12: Tổ chức y tế thế giới viết tắt là?", "options": ["WHO", "WTO", "UNICEF", "IMF"], "answer": "WHO"},
            {"question": "Câu 13: Hành tinh lớn nhất trong hệ mặt trời?", "options": ["Sao Mộc", "Sao Thổ", "Sao Hỏa", "Sao Kim"], "answer": "Sao Mộc"},
            {"question": "Câu 14: Quốc gia nào có diện tích nhỏ nhất?", "options": ["Vatican", "Monaco", "Nauru", "San Marino"], "answer": "Vatican"},
            {"question": "Câu 15: Ai phát minh ra bóng đèn điện?", "options": ["Thomas Edison", "Nikola Tesla", "Alexander Graham Bell", "James Watt"], "answer": "Thomas Edison"}
        ]
        self.current_question_index = 0
        self.score = 0
        self.wrong_attempts = 0

    def get_current_question(self):
        if self.current_question_index < len(self.questions):
            return self.questions[self.current_question_index]
        else:
            return None

    def check_answer(self, selected_option):
        correct_answer = self.questions[self.current_question_index]["answer"]
        if selected_option == correct_answer:
            self.score += 1
            return True
        else:
            self.wrong_attempts += 1
            return False

    def next_question(self):
        self.current_question_index += 1

    def reset(self):
        self.current_question_index = 0
        self.score = 0
        self.wrong_attempts = 0

class MillionaireGame:
    def __init__(self, root):
        self.root = root
        self.root.title("Ai Là Triệu Phú")
        
        # Set window size
        self.root.geometry("1300x1000")
        
        # Load background image
        self.bg_image = Image.open("bg.png")
        self.bg_image = self.bg_image.resize((1300, 700), Image.LANCZOS)
        self.bg_photo = ImageTk.PhotoImage(self.bg_image)
        
        # Create canvas for background image
        self.canvas = tk.Canvas(root, width=1300, height=700)
        self.canvas.pack(fill="both", expand=True)
        self.canvas.create_image(0, 0, image=self.bg_photo, anchor="nw")
        
        # Create buttons on the start screen
        self.start_button = tk.Button(root, text="Bắt Đầu", font=("Arial", 20), command=self.start_game)
        self.rules_button = tk.Button(root, text="Luật Chơi", font=("Arial", 20), command=self.show_rules)
        self.start_button_window = self.canvas.create_window(600, 300, anchor="nw", window=self.start_button)
        self.rules_button_window = self.canvas.create_window(600, 400, anchor="nw", window=self.rules_button)
        
        # Initialize Quiz
        self.quiz = Quiz()
    
    def start_game(self):
        # Reset game
        self.quiz.reset()
        
        # Remove start and rules buttons
        self.canvas.delete(self.start_button_window)
        self.canvas.delete(self.rules_button_window)
        
        # Setup question label and option buttons
        self.question_label = tk.Label(self.root, text="", font=("Arial", 16), wraplength=500, bg="white")
        self.question_label.pack(pady=20)
        
        self.option_buttons = []
        for i in range(4):
            button = tk.Button(self.root, text="", font=("Arial", 14), command=lambda i=i: self.check_answer(i))
            button.pack(pady=5, fill="x")
            self.option_buttons.append(button)
        
        self.next_question()
    
    def show_rules(self):
        rules = ("Luật chơi:\n"
                 "1. Trả lời các câu hỏi bằng cách chọn một trong bốn đáp án.\n"
                 "2. Mỗi câu trả lời đúng sẽ ghi được một điểm.\n"
                 "3. Mỗi người chơi có 2 lượt chọn sai.\n"
                 "4. Trò chơi kết thúc khi hết lượt chọn sai hoặc hoàn thành tất cả các câu hỏi.")
        messagebox.showinfo("Luật Chơi", rules)
    
    def next_question(self):
        current_question = self.quiz.get_current_question()
        if current_question:
            self.question_label.config(text=current_question["question"])
            for i, option in enumerate(current_question["options"]):
                self.option_buttons[i].config(text=option)
        else:
            messagebox.showinfo("Game Over", f"Bạn đã trả lời đúng {self.quiz.score} câu hỏi!")
            self.reset_game()
    
    def check_answer(self, index):
        current_question = self.quiz.get_current_question()
        if current_question:
            selected_option = current_question["options"][index]
            if self.quiz.check_answer(selected_option):
                messagebox.showinfo("Đúng", f"Câu trả lời chính xác!\nĐiểm hiện tại: {self.quiz.score}")
            else:
                remaining_attempts = 2 - self.quiz.wrong_attempts
                if remaining_attempts > 0:
                    messagebox.showerror("Sai", f"Câu trả lời sai! Bạn còn {remaining_attempts} lượt sai.")
                else:
                    messagebox.showerror("Game Over", "Bạn đã hết lượt trả lời sai!")
                    self.reset_game()
                    return
            self.quiz.next_question()
            self.next_question()
    
    def reset_game(self):
        self.question_label.pack_forget()
        for button in self.option_buttons:
            button.pack_forget()
        self.start_button_window = self.canvas.create_window(600, 300, anchor="nw", window=self.start_button)
        self.rules_button_window = self.canvas.create_window(600, 400, anchor="nw", window=self.rules_button)

if __name__ == "__main__":
    root = tk.Tk()
    game = MillionaireGame(root)
    root.mainloop()
```
